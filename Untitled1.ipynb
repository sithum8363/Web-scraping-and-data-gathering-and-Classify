```python
!pip install selenium webdriver-manager
!apt-get update
# Install Google Chrome from its official repository
!wget -q -O - https://dl-ssl.google.com/linux/linux_signing_key.pub | apt-key add -
!echo "deb [arch=amd64] http://dl.google.com/linux/chrome/deb/ stable main" | tee /etc/apt/sources.list.d/google-chrome.list
!apt-get update
!apt-get install -y google-chrome-stable
import time
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

class news:
    def __init__(self, url, driver=None):
        self.url =url
        if driver:
            self.driver = driver
            self._driver_owned_by_instance = False # Indicate that this instance does not own the driver
        else:
            # Use ChromeDriverManager to automatically handle ChromeDriver setup
            service = Service(ChromeDriverManager().install())
            options = Options()
            options.add_argument('--headless=new')
            options.add_argument('--no-sandbox')
            options.add_argument('--disable-dev-shm-usage')
            self.driver = webdriver.Chrome(service=service, options=options)
            self._driver_owned_by_instance = True # Indicate that this instance owns the driver



    def line(self):
        self.driver.get(self.url)


    def scroll(self):
      self.start=time.time()
      last_height = self.driver.execute_script("return document.body.scrollHeight")
      while True:
        self.driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        new_height = self.driver.execute_script("return document.body.scrollHeight")
        if time.time()-self.start >3600:
          break


    def get(self):
        article_links = self.driver.find_elements(By.CSS_SELECTOR, "#news-cards a.card-v4")
        text = self.driver.find_element(By.TAG_NAME, "body").text
        print(text)
        with open('filename.txt', 'w') as file:
            for link in article_links:
                href = link.get_attribute("href")
                print(f"Full URL: {href}")
                file.writelines([href,'\n'])
        # Removed self.driver.quit() from here, as driver management is handled externally
        file.close()

    def H4tage(self):
        try:
            sd=self.driver.find_element(By.TAG_NAME, "h1")
            return sd.text
        except Exception:
            return ""
    def Ptage(self):
        try:
            sd=self.driver.find_element(By.CSS_SELECTOR, ".description-content.col-md-9")
            return sd.text
        except Exception:
            return ""

    def quit_driver(self):
        if self._driver_owned_by_instance:
            self.driver.quit()

working=news('https://hirunews.lk/en/news_listing.php?category=General')
working.line()
working.scroll()
working.get()
working.quit_driver()


# Second phase: Process each article using a single shared driver
class lookoneby: # Changed to not inherit from news, as it's a processor, not a news page itself.
    def __init__(self):
        # Initialize a single shared driver for processing all articles
        service = Service(ChromeDriverManager().install())
        options = Options()
        options.add_argument('--headless=new')
        options.add_argument('--no-sandbox')
        options.add_argument('--disable-dev-shm-usage')
        options.binary_location = '/usr/bin/google-chrome'
        self.shared_driver = webdriver.Chrome(service=service, options=options)

        self.li_ac=["fire",'accident','crash','collision','mishap','tragedy','lost',"collide","Pedestrian","elephant attack"]
        self.li_ve=["vehicle","bus","lorry","train","airplane","motorcycle","overturned","derailed","runway","pile-up","Three-wheeler","Motorbike","motorbike"]
        self.li_ce=["black market","action","ICE","murder","body","robbery","theft","assault","suspect","arrested","shooting","fraud","investigation","court","sentenced","illegal","Special search operation","investigations","dead","Crimes Division","operation"]
        self.li_gv=["Commission","election","Parliament","government’s","Minister","Ministry","policy","government","law","agreement","protest","diplomatic","vote","shipment","MRIA","Authority","distributor","distributors","support","announced","restructured","CEB","stocks"]
        self.li_in=["carriers","US","Ambassador","Higher Education Scholarship","strikes","Middle-East","International","international","global","overseas","abroad","world leaders","summit","treaty","agreement","alliance","diplomatic","diplomacy","sanctions","embassyambassador","border","conflict","war","ceasefire","negotiation","peace talks","trade deal","foreign policy","bilateral","multilateral","United Nations","NATO","European Union","security council","Foreign Minister","airspace"]
        self.li_mr=["market","stock","shares","company","profit","loss","investment","trade","economy","inflation","bank","finance","revenue","growth","industry","merger","cancellations","investment"]
        self.li_his=['Poya','history','Buddha','Buddhist','Sasana','Full Moon Poya','blessing',"church","bishop","good friday","mosque","umrah","muslim","islamic","eid"]
        self.li_we=["weather","forecast","temperature","rain","rainfall","showers","drizzle","thunder","lightning","storm","thunderstorm","cloudy","sunny","wind","windy","strong winds","gusts","humidity","heatwave","cold wave","high pressure","low pressure","climate","meteorology","weather department","meteorological department","weather warning","advisory","heavy rain","flood warning","rough seas","cyclone","hurricane","typhoon","monsoon","snow","snowfall","hail","mist","fog"]
        self.li_He=["virus","disease","outbreak","hospital","medical","treatment","vaccine","infection","public health","doctor","patient","epidemic"]
        with open('filename.txt', 'r') as file,open('news_dataset.csv', 'a', newline="", encoding='utf-8') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(["url", "headline", "paragraph","type of new"])
            lines=file.readlines()
            for ui in lines:
                try:
                    # Create a news object for each URL, passing the shared driver
                    article_news = news(ui.strip(), driver=self.shared_driver) # .strip() to remove newline characters
                    article_news.line() # Navigate the shared driver to the article URL
                    text_words= article_news.Ptage()
                    head=article_news.H4tage()
                    text_word=text_words.split()
                    foundone = any(word in text_word for word in self.li_ac)
                    foundtwo = any(word in text_word for word in self.li_ve)
                    foundthree = any(word in text_word for word in self.li_ce)
                    foundfour = any(word in text_word for word in self.li_gv)
                    foundfive = any(word in text_word for word in self.li_in)
                    foundsix=any(word in text_word for word in self.li_mr)
                    foundseven=any(word in text_word for word in self.li_his)
                    foundeight=any(word in text_word for word in self.li_we)
                    foundnine=any(word in text_word for word in self.li_He)
                    if (foundone and foundtwo):
                      print("yes")
                    print(article_news.Ptage().split())
                    writer.writerow([ui, head, text_words, "accident" if (foundone and foundtwo) else "Crime" if(foundthree)else "government" if(foundfour)else "International News" if(foundfive)else "Economy News" if(foundsix)else "Religious"if(foundseven)else "wether"if(foundeight)else"HELATH"if(foundnine)else"NORMALE"])
                except Exception as e:
                    print(f"Error processing URL {ui.strip()}: {e}")
                    # If the driver crashes, try to re-initialize it to continue
                    self.shared_driver.quit()
                    service = Service(ChromeDriverManager().install())
                    options = Options()
                    options.add_argument('--headless=new')
                    options.add_argument('--no-sandbox')
                    options.add_argument('--disable-dev-shm-usage')
                    options.binary_location = '/usr/bin/google-chrome'
                    self.shared_driver = webdriver.Chrome(service=service, options=options)
                    print("Re-initialized shared driver due to error. Continuing to next URL...")

        self.shared_driver.quit()

lookoneby()
```
